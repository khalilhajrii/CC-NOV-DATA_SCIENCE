# Functions

```python
#Basic informations about the data set
def basic_informations():
    df.describe()
    #Changing variables names
    df.rename(columns = {"Date" : "date", "Open" : "open", "High" : "high", "Low" : "low","Close" : "close", "Volume" : "volume", "Currency" : "currency"}, inplace = True)
    print(df.columns)
    print("Shape of the dataset: ", df.shape)
    print("Length of the dataset: ", len(df))
    print("Maximum closing price of gold during last twenty two years: ", df["close"].max())
    print("Minimum closing price of gold during last twenty two years: ", df["close"].min())
    
#----------------------------------------------corelation-----------------------------------------------------------------------    
#datavisualisation
def datavis(data):
    f, ax = plt.subplots(figsize= [20,15])
    sns.heatmap(data.corr(), annot=True, fmt=".2f", ax=ax, cmap = "magma" )
    ax.set_title("Correlation Matrix", fontsize=20)
    plt.show()
    fig, axes = plt.subplots(1, 3, figsize = (22, 5))

#---------------------------------------------Univariate distribution-----------------------------------------------------------
    sns.distplot(ax = axes[0], x = df["open"], hist = True, bins = 10, kde = True,
             vertical = False, color = "#0E0E0D").set(title = "open price distribution");

    sns.distplot(ax = axes[1], x = df["high"], hist = True, bins = 10, kde = True,
             vertical = False, color = "#1F780D").set(title = "high price distribution");

    sns.distplot(ax = axes[2], x = df["low"], hist = True, bins = 10, kde = True,
             vertical = False, color = "#780D62").set(title = "low price distribution");
    
#---------------------------------------------Corelation using pairplot---------------------------------------------------------
    sns.pairplot(df, diag_kind = "reg", height = 2, aspect = 1, corner = True);
    
#---------------------------------------------Joinplot--------------------------------------------------------------------------
    sns.jointplot(x = "close", y = "low", data = df, height = 8, ratio = 6, kind = "scatter");

    

#Visualisation per date    
def datavistimeseries(data):
#---------------------------------------------Close and date plot---------------------------------------------------------------
    fig = plt.figure(figsize = (15, 10))
    plt.plot(df["date"], df["close"], color = "#0E7791")
    plt.title("Gold close price")
    plt.xlabel("date", fontsize = 15)
    plt.ylabel("close price of gold in USD", fontsize = 15)
    plt.show()

#---------------------------------------------open and date plot---------------------------------------------------------------
    plt.plot(df["date"], df["open"], color = "#D3C404")
    plt.title("Gold open price")
    plt.xlabel("date", fontsize = 15)
    plt.ylabel("open price of gold in USD", fontsize = 15)
    plt.show()
    
    
    
#---------------------------------------------Data preprocessing---------------------------------------------------------------
def convertingarray(d):
    #Converting new dataframe to array
    data=d.filter(['close'])
    return data

#Training data
def datalength(d):
    data_length = int(np.ceil( len(d) * .95 ))
    return data_length

#Normalizing data
def normalizing(d):
    scaler = MinMaxScaler(feature_range = (0, 1))
    scaled_data=scaler.fit_transform(d)
    return scaled_data, scaler
     
#Creating training dataset
def train_test(scaling, data_length):
    train_data = scaling[0:int(data_length), :]
    x_train = []
    y_train = []
    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i, 0])
        y_train.append(train_data[i, 0])
    return x_train, y_train

#Convert x_train and y_train to array
def array_train_test(x, y):
    x = np.array(x)
    y = np.array(y)
    return x, y

#Reshaping the data
def reshaping(x):
    x = np.reshape(x, (x.shape[0], x.shape[1], 1))
    return x

#---------------------------------------------LONG SHORT-TERM MEMORY (LSTM) Model---------------------------------------------------------------
# Building and training LSTM model
model = Sequential()
def modeling(x, y):
    model.add(LSTM(128, return_sequences = True, input_shape = (x.shape[1], 1)))
    model.add(LSTM(64, return_sequences = False))
    model.add(Dense(25))
    model.add(Dense(1))

    model.compile(optimizer = "adam", loss = "mean_squared_error")

    model.fit(x, y, batch_size = 1, epochs = 1)
    return model
    

#---------------------------------------------------Model Testing----------------------------------------------------------------------
#Spliting test data
def testing(d, scaling, data_length):
    test_data = scaling[data_length - 60: , :]
    x_test = []
    y_test = d[data_length:, :]
    for i in range(60, len(test_data)):
        x_test.append(test_data[i-60:i, 0])
    return x_test, y_test

#Convert test data to array and reshape it
def reshape_test():
    x = np.array(x)
    x = np.reshape(x, (x.shape[0], x.shape[1], 1 ))
    return x

#Get predicted prices
def prediction(x):
    predicted = model.predict(x)
    predicted = scaler.inverse_transform(predicted)
    return predicted
    
#Calculating error
def error_calculation(predicting, y):
    rmse = np.sqrt(np.mean(((predicting - y) ** 2)))
    return rmse

#Visualise output
def visualize_output(d, predicting, data_length):
    train = d[:data_length]
    valid = d[data_length:]
    valid['Predictions'] = predicting
    # Visualize the data
    plt.figure(figsize=(16,6))
    plt.title('Model')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Close Price USD ($)', fontsize=18)
    plt.plot(train['close'])
    plt.plot(valid[['close', 'Predictions']])
    plt.legend(['Train', 'Val', 'Predictions'], loc = 'lower right')
    plt.show()
    
```
